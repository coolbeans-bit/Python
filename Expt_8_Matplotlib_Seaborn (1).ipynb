# Line graph

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5,6,7]
temp = [32 , 34 , 35 , 31 , 36 , 29 , 27]
plt.plot(days , temp)
```

```python
plt.plot(days , temp)
plt.xlabel('DAYS DATA')
plt.ylabel('TEMP RECORDS')
plt.title('Weather information')
```

```python
plt.plot(days , temp , c = 'r' , marker = '>' , ls = ':')

plt.xlabel('DAYS DATA')
plt.ylabel('TEMP RECORDS')
plt.title('Weather information')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.hist([blood_sugar_men , blood_sugar_women] , bins= [77,100 , 125 , 150] 
         , label = ['Men' , 'Women'])

plt.legend()
```

```python
plt.hist([blood_sugar_men , blood_sugar_women] , bins= [77,100 , 125 , 150] 
         , label = ['Men' , 'Women'] , color = ['green' , 'magenta'] , histtype = 'stepfilled')

plt.legend()
```
