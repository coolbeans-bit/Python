# Line graph

```python
import matplotlib.pyplot as plt
```

```python
days = [1,2,3,4,5,6,7]

temp = [32 , 34 , 35 , 31 , 36 , 29 , 27]
```

```python
plt.plot(days , temp)
```

```python
plt.plot(days , temp)
plt.xlabel('DAYS DATA')
plt.ylabel('TEMP RECORDS')
plt.title('Weather information')
```

```python
plt.plot(days , temp , c = 'b' , marker = 'o' , ls = ':')
plt.xlabel('DAYS DATA')
plt.ylabel('TEMP RECORDS')
plt.title('Weather information')
```
